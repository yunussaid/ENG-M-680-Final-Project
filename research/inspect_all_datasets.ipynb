```python
import os
import zipfile
import pandas as pd
import sqlite3
```

## 1. Quality Prediction in a Mining Process

```python
# Unzip dataset only if needed
if not os.path.exists("./datasets/quality_prediction_in_a_mining_process.csv"):
    with zipfile.ZipFile("./datasets/quality_prediction_in_a_mining_process.zip", 'r') as zip_ref:
        zip_ref.extractall("./datasets")

# Load dataset
df_mining = pd.read_csv("./datasets/quality_prediction_in_a_mining_process.csv")
print('df_mining ', df_mining.shape, ':', sep='')

with pd.option_context('display.max_columns', None):
    display(df_mining.head())
```

## 2. European Soccer Database

```python
with sqlite3.connect("./datasets/european_soccer_database.sqlite") as conn:
    tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
    print('Available Tables:\n', tables, '\n')
    for table_name in tables["name"]:
        df_soccer = pd.read_sql_query(f"SELECT * FROM {table_name};", conn)
        print(f"\nTable Name: {table_name} {df_soccer.shape}")
        with pd.option_context('display.max_columns', None):
            if table_name in ['sqlite_sequence', 'League', 'Country']:
                display(df_soccer)
            else:
                display(df_soccer.head())
```

## 3. Air Quality Data in India (2015 - 2020)

```python
files = [
    './datasets/air_quality_data_in_india_city_day.csv',
    './datasets/air_quality_data_in_india_city_hour.csv',
    './datasets/air_quality_data_in_india_station_day.csv',
    './datasets/air_quality_data_in_india_station_hour.csv',
    './datasets/air_quality_data_in_india_stations.csv'
]

for file in files:
    df_air = pd.read_csv(file, low_memory=False)
    df_name = file.replace('./datasets/air_quality_data_in_india_', '').replace('.csv', '')
    print(df_name, ' ', df_air.shape, ':', sep='')
    display(df_air.head())
```
